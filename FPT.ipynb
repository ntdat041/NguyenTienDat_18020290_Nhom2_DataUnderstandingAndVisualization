```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
FPT = pd.read_csv('FPT.csv')
```

```python
FPT['Type'].value_counts()
```

```python
Laptop = FPT[FPT['Type'] == 'Máy tính xách tay']
Phone = FPT[FPT['Type'] == 'Điện thoại']
Tablet = FPT[FPT['Type'] == 'Máy tính bảng']
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(Laptop['Current Price'],Laptop['Rate'])
# set a title and labels
ax.set_title('Laptop Dataset')
ax.set_xlabel('Price')
ax.set_ylabel('Rate')
```

```python
fig, ax = plt.subplots()

ax.scatter(Phone['Current Price'],Phone['Rate'])

ax.set_title('Phone Dataset')
ax.set_xlabel('Price')
ax.set_ylabel('Rate')
```

```python
fig, ax = plt.subplots()

ax.scatter(Tablet['Current Price'],Tablet['Rate'])

ax.set_title('Tablet Dataset')
ax.set_xlabel('Price')
ax.set_ylabel('Rate')
```

```python
# create color dictionary
colors = {'Máy tính xách tay':'r', 'Điện thoại':'g',  'Máy tính bảng':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(FPT['Current Price'])):
    ax.scatter(FPT['Current Price'][i], FPT['Rate'][i],color=colors[FPT['Type'][i]])
# set a title and labels
ax.set_title('FPT Dataset')
ax.set_xlabel('Price')
ax.set_ylabel('Rate')
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = FPT['Type'].value_counts()
# get x and y data 
types = data.index
number = data.values 
# create bar chart 
ax.bar(types,number) 
# set title and labels 
ax.set_title('FPT Shop') 
ax.set_xlabel('') 
ax.set_ylabel('Số lượng')
```
